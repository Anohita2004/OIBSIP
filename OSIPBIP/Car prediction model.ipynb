```python
#Importing the libraries
import pandas as pd
import numpy as np
```

```python
#Loading the dataset
df=pd.read_csv('car data.csv')
```

```python
#Displaying the first five rows of the data set
df.head()
```

```python
#The basic information about the dataset
df.info()
```

```python
#The name of the columns of the dataset
df.columns
```

```python
#Dropping the car name column
df=df.drop(columns=['Car_Name'])
df
```

```python
#Importing the Label Encoder to encode the values under the respectively chosen  columns 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Fuel_Type']= le.fit_transform(df['Fuel_Type'])
df['Year']= le.fit_transform(df['Year'])
df['Transmission']= le.fit_transform(df['Transmission'])
```

```python
df
```

```python
#On analysing the dataset,it is clearly evident that Selling type is not affecting the selling price of the car ,hence,we drop those columns.
x=df.drop(['Selling_Price','Selling_type'],axis=1)#independent variable
y=df['Selling_Price']#dependent variable
```

```python
#Grouping the data for training the data and testing the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
#We select Linear Regression as our model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python

from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_percentage_error
#Calculating the mean squared error
mse=mean_squared_error(y_test,y_predict)
print("the mean squared error is %f"%mse)

#Calculating the r2 score
r2_score= r2_score(y_test,y_predict)
print("the r2 score is %f"%r2_score)

#Calculating the mean absolute error percentage
mape=mean_absolute_percentage_error(y_test,y_predict)
print('the mean absolute error percentage is %f'%mape)
```
